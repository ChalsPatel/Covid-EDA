```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datasets
from pandas.plotting import scatter_matrix
```

```python
prefix_url = "https://graphemy.ir/datasets"


def get_file(name):
    file_name = name if "." in name else f"{name}.csv"
    return f"{prefix_url}/{file_name}"

df = pd.read_excel(get_file('Masih_Data.xlsx'),skiprows=1)
df.head()
```

Searching if NaN values exist in our data:

```python
df.isna().sum()
```

Fill Nan data:

```python
df.fillna('filled', inplace=True)
df.isna().sum()
```

Removing duplicated data:

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

The df.info() method provides information about the data types and the number of non-null entries for each column.


```python
df.info()
```

The df.describe() method provides summary statistics like mean, standard deviation, minimum, and maximum values for numerical columns.

```python
df.describe()
```

the distribution of values within each column.

```python
df.hist(figsize=(15,17))
plt.suptitle('Histograms of numerical columns')
plt.show()
```

The scatter plot of 'age' vs 'HTN' suggests that there is an interest in understanding how age correlates with hypertension (HTN).

```python
plt.figure(figsize=(8,6))
plt.scatter(df['age'], df['HTN'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('HTN')
plt.title('Age vs HTN')

plt.show()
```

The scatter matrix provides a comprehensive view of pairwise relationships between all numerical variables, helping to identify potential correlations.

```python
scatter_matrix(df, figsize=(20,20), alpha=0.5, diagonal='kde')
plt.suptitle('Scatter matrix of numerical columns')
plt.show()
```

```python
selected_columns = df[['Outcome', 'age', 'HTN','BP max 1', 'BP min 1' ]]
scatter_matrix(selected_columns, figsize=(10,10), alpha=0.5, diagonal='kde')
plt.suptitle('Scatter matrix of selected numerical columns')
plt.show()
```

```python
selected_columns = df[['sex', 'DM', 'age']]
scatter_matrix(selected_columns, figsize=(10,10), alpha=0.5, diagonal='kde')
plt.suptitle('Scatter matrix of selected numerical columns')
plt.show()
```
